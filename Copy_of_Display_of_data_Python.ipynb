```python
import numpy as np
import math
t = np.arange(0,8000,1)
xf1 =  np.sin((2*np.pi*50)*(1/8000)*t)
xf2 =  np.sin((2*np.pi*50)*(1/8000)*t+180)
xf3 =  np.sin((2*np.pi*50)*(1/8000)*t-180)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

fig, ax = plt.subplots(3,1,figsize=(15,10))
plt.rcParams.update({'font.size': 15})

# Set axis ranges; by default this will put major ticks every 25.
ax[0].set_xlim(0, 1000)
ax[0].set_ylim(-1.2, 1.2)

# Change major ticks to show every 20.
ax[0].xaxis.set_major_locator(MultipleLocator(100))
ax[0].yaxis.set_major_locator(MultipleLocator(0.4))

# Change minor ticks to show every 5. (20/4 = 5)
ax[0].xaxis.set_minor_locator(AutoMinorLocator(4))
ax[0].yaxis.set_minor_locator(AutoMinorLocator(4))

# Turn grid on for both major and minor ticks and style minor slightly
# differently.
ax[0].grid(which='major', color='#CCCCCC', linestyle='--')
ax[0].grid(which='minor', color='#CCCCCC', linestyle=':')

ax[0].plot(t,xf1,color='b')

ax[0].set_title('Fase (1)',style = "italic",fontsize=23)
ax[0].set_xlabel('\n'r'$(a)$')
ax[0].legend([r'$ F_{1} $'],ncol=3,loc='upper right',fontsize=23,fancybox=True, framealpha=0.5)


# Set axis ranges; by default this will put major ticks every 25.
ax[1].set_xlim(0, 1000)
ax[1].set_ylim(-1.2, 1.2)

# Change major ticks to show every 20.
ax[1].xaxis.set_major_locator(MultipleLocator(100))
ax[1].yaxis.set_major_locator(MultipleLocator(0.4))

# Change minor ticks to show every 5. (20/4 = 5)
ax[1].xaxis.set_minor_locator(AutoMinorLocator(4))
ax[1].yaxis.set_minor_locator(AutoMinorLocator(4))


# Turn grid on for both major and minor ticks and style minor slightly
# differently.
ax[1].grid(which='major', color='#CCCCCC', linestyle='--')
ax[1].grid(which='minor', color='#CCCCCC', linestyle=':')

ax[1].plot(t,xf2,color='r')
ax[1].set_title('Fase (2)',style = "italic",fontsize=23)
ax[1].set_xlabel('\n' r'$(b)$')
ax[1].legend([r'$ F_{2} $'],ncol=3,loc='upper right',fontsize=23,fancybox=True, framealpha=0.5)


# Set axis ranges; by default this will put major ticks every 25.
ax[2].set_xlim(0, 1000)
ax[2].set_ylim(-1.2, 1.2)

# Change major ticks to show every 20.
ax[2].xaxis.set_major_locator(MultipleLocator(100))
ax[2].yaxis.set_major_locator(MultipleLocator(0.4))

# Change minor ticks to show every 5. (20/4 = 5)
ax[2].xaxis.set_minor_locator(AutoMinorLocator(4))
ax[2].yaxis.set_minor_locator(AutoMinorLocator(4))


# Turn grid on for both major and minor ticks and style minor slightly
# differently.
ax[2].grid(which='major', color='#CCCCCC', linestyle='--')
ax[2].grid(which='minor', color='#CCCCCC', linestyle=':')

ax[2].plot(t,xf3,color='g')
ax[2].set_title('Fase (3)',style = "italic",fontsize=23)
ax[2].set_xlabel('Tiempo [Segundos]\n' r'$(c)$')
ax[2].legend([r'$ F_{3} $'],ncol=3,loc='upper right',fontsize=23,fancybox=True, framealpha=0.5)
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

fig, ax = plt.subplots(1,1,figsize=(10,5))
plt.rcParams.update({'font.size': 15})

# Set axis ranges; by default this will put major ticks every 25.
ax.set_xlim(0, 1000)
ax.set_ylim(-1.1, 1.1)

# Change major ticks to show every 20.
ax.xaxis.set_major_locator(MultipleLocator(100))
ax.yaxis.set_major_locator(MultipleLocator(0.2))

# Change minor ticks to show every 5. (20/4 = 5)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))

# Turn grid on for both major and minor ticks and style minor slightly
# differently.
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

ax.plot(t,xf1,color='b')
ax.plot(t,xf2,color='r')
ax.plot(t,xf3,color='g')

ax.set_title('Grafica Trifasica',style = "italic",fontsize=23)
ax.set_xlabel('Tiempo [Segundos]')
ax.legend([r'$ F_{1} $',r'$ F_{2} $',r'$ F_{3} $'],ncol=3,loc='upper right',fontsize=23,fancybox=True, framealpha=0.5)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes,mark_inset
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig, ax = plt.subplots(1,1,figsize=(20,10))
plt.rcParams.update({'font.size': 23})

ones1 = np.ones((len(t),1))

# Set axis ranges; by default this will put major ticks every 25.
ax.set_xlim(0, 1000)
ax.set_ylim(-1.1, 1.1)

# Change major ticks to show every 20.
ax.xaxis.set_major_locator(MultipleLocator(200))
ax.yaxis.set_major_locator(MultipleLocator(0.2))

# Change minor ticks to show every 5. (20/4 = 5)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))


# Turn grid on for both major and minor ticks and style minor slightly differently.
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

ax.plot(t,xf1,color='b')
ax.plot(t,xf2,color='r')
ax.plot(t,xf3,color='g')
ax.set_title('Graficas de Fase con Zoom',style = "italic",fontsize=23)
ax.set_ylabel('Presion [Bar]')
ax.set_xlabel('Tiempo [Horas]\n' r'$(b)$')
ax.legend([r'$ F_{1} $',r'$ F_{2} $',r'$ F_{3} $'],ncol=7,loc='upper right',fontsize=23,fancybox=True, framealpha=0.5)
plt.tight_layout()


### Primer Zoom
axins = zoomed_inset_axes(ax,2,loc='upper center',
                          axes_kwargs={"facecolor" : "white"},borderpad=3)

axins.plot(t,xf1,color='b')
axins.plot(t,xf2,color='r')
axins.plot(t,xf3,color='g')

# x1,x2,y1,y2 = 180, 190, 0.04, 0.06
x1,x2,y1,y2 = 100, 200, 0.3, 0.6
axins.set_xlim(x1,x2)
axins.set_ylim(y1,y2)

# Change major ticks to show every 20.
axins.xaxis.set_major_locator(MultipleLocator(50))
axins.yaxis.set_major_locator(MultipleLocator(0.1))

# Change minor ticks to show every 5. (20/4 = 5)
axins.xaxis.set_minor_locator(AutoMinorLocator(4))
axins.yaxis.set_minor_locator(AutoMinorLocator(4))

# Turn grid on for both major and minor ticks and style minor slightly
# differently.
axins.grid(which='major', color='#CCCCCC', linestyle='--')
axins.grid(which='minor', color='#CCCCCC', linestyle=':')

pp,p1,p2 = mark_inset(ax,axins,loc1=2,loc2=3,alpha=1)
pp.set_fill(True)

# pp.spines['top'].set_linewidth(3)
pp.set_facecolor("gray")
pp.set_edgecolor("black")
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 15)


### Segundo Zoom
axins1 = zoomed_inset_axes(ax,2,loc='right', bbox_to_anchor=(1800,300),
                        axes_kwargs={"facecolor" : "white"})

axins1.plot(t,xf1,color='b')
axins1.plot(t,xf2,color='r')
axins1.plot(t,xf3,color='g')

x1,x2,y1,y2 = 700, 800, 0.5, 0.8
axins1.set_xlim(x1,x2)
axins1.set_ylim(y1,y2)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 15)

# Change major ticks to show every 20.
axins1.xaxis.set_major_locator(MultipleLocator(100))
axins1.yaxis.set_major_locator(MultipleLocator(0.1))

# Change minor ticks to show every 5. (20/4 = 5)
axins1.xaxis.set_minor_locator(AutoMinorLocator(4))
axins1.yaxis.set_minor_locator(AutoMinorLocator(4))

# Turn grid on for both major and minor ticks and style minor slightly
# differently.
axins1.grid(which='major', color='#CCCCCC', linestyle='--')
axins1.grid(which='minor', color='#CCCCCC', linestyle=':')

pp,p1,p2 = mark_inset(ax,axins1,loc1=2,loc2=1,alpha=1)
pp.set_fill(True)
# pp.spines['top'].set_linewidth(3)
pp.set_facecolor("gray")
pp.set_edgecolor("black")
```
